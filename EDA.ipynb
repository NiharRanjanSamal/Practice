```python
import kagglehub
import pandas as pd
import os

# Download latest version (will cache it)
path = kagglehub.dataset_download("yashdevladdha/uber-ride-analytics-dashboard")
print("Path to dataset files:", path)

# List files inside dataset
files = os.listdir(path)
print("Files in dataset:", files)

# Example: load a CSV file into pandas
csv_file = os.path.join(path, "ncr_ride_bookings.csv")  # change filename if different
df = pd.read_csv(csv_file)

print(df.head())
```

```python
booking_bins = pd.cut(df['Booking Value'], bins=10)
distance_bins = pd.cut(df['Ride Distance'], bins=10)

# Create crosstab
cross_tab = pd.crosstab(booking_bins, distance_bins)

plt.figure(figsize=(10,6))
sns.heatmap(cross_tab, cmap="YlGnBu")
plt.title("Heatmap of Booking Value vs Ride Distance (Binned)")
plt.show()
```
